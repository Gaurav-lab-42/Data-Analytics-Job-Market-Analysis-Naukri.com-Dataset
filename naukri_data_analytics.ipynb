```python
# Install required packages (only needed once)
!pip install pandas matplotlib seaborn
!pip install matplotlib
import matplotlib.pyplot as plt
print("Matplotlib installed successfully!")
import sys
print(sys.executable)





# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv(r"C:\Users\91935\Downloads\NaukriData_data analytics.csv")

# Show first 5 rows
print(df.head())

# Check column names and info
print("\nColumn Names:", df.columns.tolist())
df.info()

# Print columns exactly to detect any invisible characters
for col in df.columns:
    print(repr(col))  # Helpful for debugging weird spaces or invisible chars

# Replace column names here with the exact ones from the print above
df = df.dropna(subset=["Job_Titles", "Locations"])

# Display updated DataFrame
print(df.head())

df["Locations"] = df["Locations"].str.title()
df["Job_Titles"] = df["Job_Titles"].str.title()

```

```python
print("Unique Job Titles:", df["Job_Titles"].nunique())
print("Unique Locations:", df["Locations"].nunique())
top_locations = df["Locations"].value_counts().head(10)
print(top_locations)
top_jobs = df["Job_Titles"].value_counts().head(10)
print(top_jobs)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.values, y=top_locations.index, palette='viridis')
plt.title("Top 10 Job Locations")
plt.xlabel("Number of Jobs")
plt.ylabel("Location")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette='coolwarm')
plt.title("Top 10 Job Titles")
plt.xlabel("Number of Jobs")
plt.ylabel("Job Title")
plt.show()
```

```python
top_locations.plot.pie(autopct='%1.1f%%', figsize=(6,6), title="Job Distribution by Location")
plt.ylabel("")  # remove y-label
plt.show()
```

```python
heatmap_data = df.pivot_table(index="Locations", columns="Job_Titles", aggfunc="size", fill_value=0)
sns.heatmap(heatmap_data)

```
